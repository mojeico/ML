```python
class User:
    def __init__(self, count = 0):
        self.count = count
        print(count)
        
        
    def click(self, num_time = 1):
        self.count += num_time;
        
        
    def getClick(self):
        return self.count
    
    def reset(self):
        self.count = 0
```

```python
# Gradient of y with respect to x
gradient_of_y = lambda x: 2*(x-5)

#Initial value of x close to zero
current_x_value = 0.1

# maximum number of iterations that can be done
maximum_iterations = 5000

# Learning Rate
learning_rate = 0.01
current_iteration = 0
previous_step_size = 0.5

precision = 0.00000001 #This tells us when to stop the algorithm

```

```python
while previous_step_size > precision and current_iteration < maximum_iterations:
    previous_x_value = current_x_value 
    current_x_value = current_x_value - learning_rate * gradient_of_y(previous_x_value)
    previous_step_size = abs(current_x_value - previous_x_value)
    current_iteration = current_iteration + 1 
    
    print("previous_step_size ",previous_step_size ,"precision",precision) 


    
    print("Iteration count ",current_iteration ,"The current value of x is",current_x_value)
    print()
    
print("The local minima found at ", current_x_value)
```
